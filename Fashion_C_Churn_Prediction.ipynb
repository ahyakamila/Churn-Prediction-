Normalisasi

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df)
```

```python
df
```

```python
scaled
```

```python
df_scaled = pd.DataFrame(scaled, columns =['Unnamed: 0',	'customer_id',	'recency',	'frequency',	'value',	'age',	'DidBuy'])
```

```python
df_scaled
```

Partisi data

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_scaled[["recency",	"frequency",	"value",	"age"]]
y = df_scaled["DidBuy"]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, shuffle=True, random_state=0)
```

Model Naive Bayes (Baseline Model)

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb_pred = gnb.fit(X_train, y_train).predict(X_test)
```

```python
gnb_pred
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
accuracy = accuracy_score(y_test, gnb_pred)
precision = precision_score(y_test, gnb_pred)
recall = recall_score(y_test, gnb_pred)
```

```python
print('accuracy of Naive Bayes :', accuracy*100)
print('precision of Naive Bayes :', precision*100)
print('recall of Naive Bayes :', recall*100)
```

Model Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion = 'gini', max_depth = 10, random_state=0)
dt_pred = dt.fit(X_train, y_train).predict(X_test)
```

```python
dt_pred
```

```python
accuracy_dt = accuracy_score(y_test, dt_pred)
precision_dt = precision_score(y_test, dt_pred)
recall_dt = recall_score(y_test, dt_pred)
```

```python
print("accuracy of DecisionTree :", accuracy_dt*100)
print("precision of DecisionTree :", precision_dt*100)
print("recall of DecisionTree :", recall_dt*100)
```

Tuning Hyperparameter of Decision Tree Model

```python
for i in range (10, 51, 5):
  for j in ["gini", "entropy"]:
    for k in ["auto", "sqrt"]:
      for l in ["best", "random"]:
        dt = DecisionTreeClassifier(max_depth=(i), criterion=(j), max_features =(k), splitter = (l))
        dt.fit(X_train, y_train)
        print('max_depth', i)
        print('criterion', j)
        print('max_features', k)
        print('splitter', l)
        dt.score(X_test, y_test)
        print(dt.score(X_test, y_test))
        print()
```

```python
dt_tuned = DecisionTreeClassifier(max_depth = 15, criterion = 'gini', max_features = 4, random_state=42, splitter = 'best')
```

```python
dt_tuned.fit(X_train, y_train)
```

```python
dt_tuned.score(X_test, y_test)
```

Cross Validation

```python
from sklearn.model_selection import cross_val_score
clf = DecisionTreeClassifier(max_depth = 15, criterion = 'gini', max_features = 4, random_state=42, splitter = 'best')
scores = cross_val_score(clf, X, y, cv = 5)
scores
```
